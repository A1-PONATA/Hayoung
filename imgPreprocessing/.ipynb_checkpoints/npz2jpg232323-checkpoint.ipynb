```python
import os
import cv2
def jpg2npz(origin_path):
    folderList = ['fp','lr','rl']
    label = {'fp':0, 'lr':1, 'rl':2}
    idx=0
    for f in folderList:
        p = origin_path + '/'+f
        
        movie_data= None #= np.empty((180, 240, 1))
        cnt=0
        for original_file in os.listdir(p):
            #print(original_file)
            file_path = p+'/'+original_file
            #img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            shrink = cv2.resize(img,(240,180))
            shrink=np.reshape(shrink,(-1,240, 180, 3))
            movie_data=np.vstack((movie_data,shrink))
            print(movie_data.shape)
            cnt+=1
           
        encoding= np.eye(3)[label[f]]
        movie_data=np.array(movie_data)
    
        print(movie_data.shape, encoding.shape)
        np.savez('/home/pirl/Documents/motion_1214ver_GRAYSCALE/'+str(30000+idx)+'.npz', train=movie_data, trining_label=encoding)
        idx+=1
```
